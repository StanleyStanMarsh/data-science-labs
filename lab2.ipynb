### Задание №5*

Выведите список (в виде объекта list) уникальных авторов за 2011 год, чье количество отзывовов хотя бы на одну книгу было больше 75 квантиля.

```python
df_amazon_2011 = df_amazon[df_amazon['Year'] == 2011]
qntl = df_amazon_2011['Reviews'].quantile(0.75)
df_amazon_q = df_amazon_2011[df_amazon_2011['Reviews'] > qntl]
authors = df_amazon_q['Author'].unique().tolist()
authors
```

### Задание №6

Создайте столбец genre_categories содержащий данные из столбца Genre, преобразованным в тип category, а затем создайте столбец genre_codes, содержащий коды категорий.

```python
df_amazon['genre_categories'] = df_amazon['Genre'].astype('category')
genre_codes_list = df_amazon['Genre'].unique().tolist()
df_amazon['genre_codes'] = [genre_codes_list.index(code) for code in df_amazon['Genre']]
df_amazon[['genre_categories', 'genre_codes']]
```
